```python
import matplotlib.pyplot as plt
labels = 'frogs','hogs','dogs','logs'
sizes = 15,20,45,15
colors = 'yellowgreen','gold','lightskyblue','lightcoral'
explode=0,0.1,0,0
plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct='%1.1f%%')
plt.axis('equal')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
#from pylab import *
 
x = np.arange(-5.0, 5.0, 0.02)
y1 = np.sin(x)
 
plt.figure(1)
plt.subplot(211)
plt.plot(x, y1)
 
plt.subplot(212)
#设置x轴范围
plt.xlim(-6, 6)
#设置y轴范围
plt.ylim(-1.5, 1.5)
plt.plot(x, y1)
plt.show()
```

```python
plt.rcParams['font.sans-serif']=['SimHei']   # 显示中文汉字
t = np.arange(0.,5.,0.2)
plt.plot(t,t,'r--',label='直线')

plt.plot(t,t**2,'b<',label='二次曲线')
plt.plot(t,t**3,'g^',label='三次函数')
plt.legend()  #显示 图例
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(1)                # 第一张图
plt.subplot(211)             # 第一张图中的第一张子图
plt.plot([1,2,3])
plt.subplot(212)             # 第一张图中的第二张子图
plt.plot([4,5,6]) 
plt.figure(2)                # 第二张图
plt.plot([4,5,6])            # 默认创建子图subplot(111)
 
plt.figure(1)                # 切换到figure 1 ; 子图subplot(212)仍旧是当前图   选定 操作的子图
plt.subplot(211)             # 令子图subplot(211)成为figure1的当前图
plt.title('Easy as 1,2,3')   # 添加subplot 211 的标题
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
 
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)
 
# 数据的直方图
n, bins, patches = plt.hist(x, 50, normed=1, facecolor='g', alpha=0.75)
 
plt.xlabel('Smarts')
plt.ylabel('Probability')
#添加标题
plt.title('Histogram of IQ')
#添加文字
plt.text(60, .025, r'$mu=100, sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```

```python
plt.rcParams['axes.unicode_minus']=False #用来正常显示负号
ax = plt.subplot(111)
 
t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2*np.pi*t)
line, = plt.plot(t, s, lw=2)
 
plt.annotate('local max', xy=(2, 1), xytext=(3, 1.5),
            arrowprops=dict(facecolor='black', shrink=0.1),
            )
 
plt.ylim(-2,2)
plt.grid(True)
plt.show()
```

```python
# 导入 matplotlib 的所有内容（nympy 可以用 np 这个名字来使用）
#from pylab import *
 
# 创建一个 8 * 6 点（point）的图，并设置分辨率为 80
plt.figure(figsize=(8,6), dpi=80)
 
# 创建一个新的 1 * 1 的子图，接下来的图样绘制在其中的第 1 块（也是唯一的一块）
plt.subplot(1,1,1)
 
X = np.linspace(-np.pi, np.pi, 256,endpoint=True)
C,S = np.cos(X), np.sin(X)
 
# 绘制余弦曲线，使用蓝色的、连续的、宽度为 1 （像素）的线条
plt.plot(X, C, color="blue", lw=2.5, ls="-",label="cosine")
 
# 绘制正弦曲线，使用绿色的、连续的、宽度为 1 （像素）的线条
plt.plot(X, S, color="r", lw=2.5, ls="--",label="sine")
 
plt.axis([-4,4,-1.2,1.2])
# 设置轴记号
 
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],[r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
 
plt.yticks([-1, 0, +1],[r'$-1$', r'$0$', r'$+1$'])
legend(loc='upper left')
# 在屏幕上显示
plt.show()
```
